```python
# Import libraries and modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import os
import seaborn as sns
```

## In this notebook I will explore the image data I will use to train my model, as well as the creation of equation data

### Look at Number of Files in Training Data

```python
# Create dictionary of label names and number of files
img_dir = '../image_data/training_images/'
file_dict = {dir:[] for dir in sorted(os.listdir(img_dir))}
for dir in sorted(os.listdir(img_dir)):
    count = 0
    for file in os.listdir(os.path.join(img_dir, dir)):
        count += 1
    file_dict[dir] = count
```

```python
file_dict
```

```python
# Create barchart of file numbers
names = list(file_dict.keys())
values = list(file_dict.values())

plt.barh(range(len(file_dict)), values, tick_label=names, edgecolor='black')
plt.title('Number of Files in Training Images')
plt.show()
```
